```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

openai_api_key = os.environ["OPENAI_API_KEY"]
client = OpenAI(api_key=openai_api_key)
```

```python
system_message = """
당신은 주어진 텍스트를 바탕으로 튜토리얼을 생성하는 AI 어시스턴트 입니다.
텍스트에 어떤 절차를 진행하는 방법에 대한 지침이 포함되어 있다면,
글머리 기호 목록 형식으로 튜토리얼을 생성하십시오.
그렇지 않으면 '텍스트에 지침이 포함되어 있지 않음'이라고 사용자에게 알리십시오.
"""

user_input = """
라면 봉지를 개봉하고, 끓는 물에 라면을 넣습니다.
약 3분 정도 시간이 지나면 빈그릇을 준비해서 최대한 가운데에 스프틀 모아 놓습니다.
그리고 방금 끓인 라면의 면만 꺼내서 스프 위에 놓습니다.
그리고 면을 끓인 뜨거운 물을 부어줍니다.
먹을 때는 젓가락으로 저어주면 라면 스프가 녹으면서 맛있는 라면이 됩니다.
단, 이 방법은 맵지 않은 라면에만 적용됩니다.
"""
```

```python
response = client.responses.create(
    model="gpt-4.1-nano",
    instructions=system_message,
    input=user_input
)

print(response.output_text)
```
